```python
import numpy as np
from sklearn.preprocessing import normalize
import scipy
import matplotlib.pyplot as plt
```

```python
th = np.linspace(0, np.pi, 100)
x = np.linspace(0, 10, 10)
plt.figure(figsize=(10,10))
for x in x:
    f = np.cos(x*np.sin(th))
    plt.plot(th,f,label='x = %d' % x)
plt.legend()
plt.show()
```
